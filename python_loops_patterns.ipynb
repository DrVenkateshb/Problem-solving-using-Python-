# 🔁 PART 1: While Loop Programs
<a id='while-loop'></a>

> A **`while` loop** runs a block of code **as long as a condition is True**. It is used when the number of iterations is not known in advance.
>
> **Syntax:**
> ```python
> while condition:
>     # body
> ```

---

## ✅ Program 1: Print Numbers from 1 to 10

**Explanation:**  
We initialize `i = 1`. The loop runs while `i <= 10`, printing `i` each time and incrementing it by 1. This is the most basic use of a `while` loop — controlled iteration using a counter variable.

````python
i = 1
while i <= 10:
    print(i)
    i += 1
````

## ✅ Program 2: Sum of First N Natural Numbers

**Explanation:**  
We accumulate a running `total` by adding each number from 1 to `n`. The loop continues until `i > n`. This demonstrates how a while loop can be used to aggregate values.

````python
n = 10
i = 1
total = 0
while i <= n:
    total += i
    i += 1
print(f"Sum of first {n} natural numbers = {total}")
````

## ✅ Program 3: Multiplication Table

**Explanation:**  
Given a number `num`, we print its multiplication table from 1 to 10. The counter `i` acts as the multiplier. This shows how loops automate repetitive arithmetic tasks.

````python
num = 5
i = 1
print(f"Multiplication Table of {num}:")
while i <= 10:
    print(f"{num} x {i} = {num * i}")
    i += 1
````

## ✅ Program 4: Factorial of a Number

**Explanation:**  
Factorial of `n` = n × (n−1) × ... × 1. We multiply `fact` by each number from `n` down to 1 using a while loop. This is a classic example of accumulative multiplication.

````python
n = 6
fact = 1
while n > 0:
    fact *= n
    n -= 1
print(f"Factorial = {fact}")
````

## ✅ Program 5: Reverse a Number

**Explanation:**  
We extract the last digit using `% 10`, build the reversed number by shifting digits left (`rev * 10 + digit`), and remove the last digit using `// 10`. The loop runs until all digits are processed.

````python
num = 12345
rev = 0
original = num
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10
print(f"Reverse of {original} = {rev}")
````

## ✅ Program 6: Check Armstrong Number

**Explanation:**  
An Armstrong number equals the sum of its digits each raised to the power of the number of digits (e.g., 153 = 1³ + 5³ + 3³). We extract digits in the loop and compute the sum of cubes.

````python
num = 153
original = num
n_digits = len(str(num))
arm_sum = 0
while num > 0:
    digit = num % 10
    arm_sum += digit ** n_digits
    num //= 10
if arm_sum == original:
    print(f"{original} is an Armstrong number")
else:
    print(f"{original} is NOT an Armstrong number")
````

## ✅ Program 7: Count Digits in a Number

**Explanation:**  
We repeatedly divide the number by 10 and count how many times we can do it before reaching 0. Each division removes one digit, so the count equals the total number of digits.

````python
num = 987654
count = 0
while num > 0:
    num //= 10
    count += 1
print(f"Number of digits = {count}")
````

## ✅ Program 8: Sum of Digits

**Explanation:**  
We extract the last digit using `% 10` and add it to `digit_sum` in each iteration. After processing all digits, we print the total sum. This is useful in checksum calculations.

````python
num = 4567
digit_sum = 0
while num > 0:
    digit_sum += num % 10
    num //= 10
print(f"Sum of digits = {digit_sum}")
````

## ✅ Program 9: Fibonacci Series

**Explanation:**  
The Fibonacci series starts with 0 and 1; each subsequent term is the sum of the previous two. We use two variables `a` and `b`, updating them in each iteration: `a, b = b, a + b`.

````python
n = 10
a, b = 0, 1
count = 0
print(f"First {n} Fibonacci numbers:")
while count < n:
    print(a, end=" ")
    a, b = b, a + b
    count += 1
````

## ✅ Program 10: Check Palindrome Number

**Explanation:**  
A palindrome number reads the same forwards and backwards (e.g., 121). We reverse the number using a while loop and compare it to the original.

````python
num = 121
original = num
rev = 0
while num > 0:
    rev = rev * 10 + num % 10
    num //= 10
if rev == original:
    print(f"{original} is a Palindrome")
else:
    print(f"{original} is NOT a Palindrome")
````

## ✅ Program 11: GCD of Two Numbers

**Explanation:**  
The **Euclidean Algorithm** finds GCD by repeatedly replacing `(a, b)` with `(b, a % b)` until `b` becomes 0. The last non-zero value of `a` is the GCD.

````python
a, b = 48, 18
while b != 0:
    a, b = b, a % b
print(f"GCD = {a}")
````

## ✅ Program 12: Print Even Numbers up to N

**Explanation:**  
We start at 2 and increment by 2 each iteration, printing only even numbers. This avoids checking every number with `%` by directly stepping through evens.

````python
n = 20
i = 2
print(f"Even numbers up to {n}:")
while i <= n:
    print(i, end=" ")
    i += 2
````

## ✅ Program 13: Power of a Number (Without Built-in)

**Explanation:**  
We compute `base^exp` by multiplying `result` by `base` repeatedly, `exp` times. This mimics the `**` operator manually and shows iterative multiplication.

````python
base, exp = 3, 4
result = 1
i = 0
while i < exp:
    result *= base
    i += 1
print(f"{base}^{exp} = {result}")
````

## ✅ Program 14: Guess the Number Game

**Explanation:**  
The loop keeps prompting the user until they guess the secret number. `break` exits the loop when the guess is correct. This is an interactive use of `while True` with a conditional break.

````python
secret = 7
print("Guess the number between 1 and 10:")
while True:
    guess = int(input("Your guess: "))
    if guess == secret:
        print("🎉 Correct!")
        break
    elif guess < secret:
        print("Too low! Try again.")
    else:
        print("Too high! Try again.")
````

## ✅ Program 15: Print Odd Numbers up to N

**Explanation:**  
Starting from 1, we jump by 2 in each step to land only on odd numbers. This is efficient since we avoid even numbers entirely without using modulus.

````python
n = 19
i = 1
print(f"Odd numbers up to {n}:")
while i <= n:
    print(i, end=" ")
    i += 2
````

## ✅ Program 16: Count Vowels in a String

**Explanation:**  
We traverse the string character by character using an index `i`. For each character, we check if it belongs to the set of vowels. This demonstrates while loop use with string indexing.

````python
s = "Hello World Python"
vowels = "aeiouAEIOU"
i = 0
count = 0
while i < len(s):
    if s[i] in vowels:
        count += 1
    i += 1
print(f"Number of vowels = {count}")
````

## ✅ Program 17: Print Numbers in Reverse

**Explanation:**  
We start `i` at `n` and decrement it by 1 each time, printing numbers in descending order. This shows how while loops can count backwards with simple condition reversal.

````python
n = 10
i = n
print(f"Numbers from {n} to 1:")
while i >= 1:
    print(i, end=" ")
    i -= 1
````

## ✅ Program 18: Check Prime Number

**Explanation:**  
We check if any number from 2 to `n-1` divides `n` evenly. If such a divisor exists, `n` is not prime. We use a `flag` variable to track divisibility and a while loop to iterate through potential divisors.

````python
n = 29
i = 2
is_prime = True
while i < n:
    if n % i == 0:
        is_prime = False
        break
    i += 1
if is_prime and n > 1:
    print(f"{n} is a Prime number")
else:
    print(f"{n} is NOT a Prime number")
````

## ✅ Program 19: ATM PIN Validator

**Explanation:**  
The loop gives the user 3 attempts to enter the correct PIN. A counter tracks failed attempts. If all 3 are wrong, the card is blocked. This demonstrates real-world usage of while loops with counters and conditions.

````python
correct_pin = "1234"
attempts = 0
while attempts < 3:
    pin = input("Enter PIN: ")
    if pin == correct_pin:
        print("✅ Access Granted!")
        break
    else:
        attempts += 1
        print(f"❌ Wrong PIN. {3 - attempts} attempt(s) left.")
else:
    print("🚫 Card blocked after 3 wrong attempts.")
````

## ✅ Program 20: Collatz Conjecture

**Explanation:**  
The Collatz Conjecture states: if `n` is even, divide by 2; if odd, multiply by 3 and add 1. Repeat until `n = 1`. The number of steps is counted. No one has proven it doesn't terminate — yet!

````python
n = 27
steps = 0
print(f"Collatz sequence starting from {n}:")
while n != 1:
    print(n, end=" → ")
    if n % 2 == 0:
        n //= 2
    else:
        n = 3 * n + 1
    steps += 1
print(f"1\nTotal steps: {steps}")
````

# 🔂 PART 2: For Loop Programs
<a id='for-loop'></a>

> A **`for` loop** iterates over a **sequence** (list, tuple, string, range, etc.). It is used when the number of iterations is **known** in advance.
>
> **Syntax:**
> ```python
> for variable in sequence:
>     # body
> ```

---

## ✅ Program 1: Print Numbers 1 to 10

**Explanation:**  
`range(1, 11)` generates numbers from 1 to 10 (11 is excluded). The `for` loop automatically handles iteration without manual incrementing — cleaner and less error-prone than while loops for known ranges.

````python
for i in range(1, 11):
    print(i, end=" ")
````

## ✅ Program 2: Sum of a List

**Explanation:**  
We iterate directly over the elements of the list without needing an index. This is idiomatic Python — `for` loops can traverse any iterable directly, making the code cleaner and more readable.

````python
numbers = [10, 20, 30, 40, 50]
total = 0
for num in numbers:
    total += num
print(f"Sum = {total}")
````

## ✅ Program 3: Find Maximum in a List

**Explanation:**  
We initialize `max_val` with the first element and compare it with each subsequent element. If a larger value is found, we update `max_val`. This is the classic linear search for maximum.

````python
numbers = [3, 41, 12, 9, 74, 15]
max_val = numbers[0]
for num in numbers:
    if num > max_val:
        max_val = num
print(f"Maximum = {max_val}")
````

## ✅ Program 4: Reverse a String

**Explanation:**  
We iterate through each character of the string and prepend it to `rev`. Since we prepend (add to front), the result is the original string in reverse order. This is a character-level string manipulation.

````python
s = "Python"
rev = ""
for ch in s:
    rev = ch + rev
print(f"Reverse of '{s}' = '{rev}'")
````

## ✅ Program 5: Count Frequency of Elements

**Explanation:**  
We use a dictionary to count occurrences. For each element, `dict.get(key, 0)` returns the current count (0 if new), and we increment it by 1. This is a fundamental frequency counting pattern in Python.

````python
items = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']
freq = {}
for item in items:
    freq[item] = freq.get(item, 0) + 1
print("Frequency:", freq)
````

## ✅ Program 6: Print All Prime Numbers up to N

**Explanation:**  
Nested for loops are used — the outer loop picks a candidate number, the inner loop checks divisibility. The `else` clause of a for loop runs when no `break` occurred, meaning no divisor was found — so it's prime!

````python
n = 50
print(f"Prime numbers up to {n}:")
for num in range(2, n + 1):
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            break
    else:
        print(num, end=" ")
````

## ✅ Program 7: Flatten a Nested List

**Explanation:**  
We use nested for loops — the outer loop iterates over each sublist, the inner loop iterates over each element in the sublist. This is a common data processing pattern called list flattening.

````python
nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
flat = []
for sublist in nested:
    for item in sublist:
        flat.append(item)
print("Flattened list:", flat)
````

## ✅ Program 8: List Comprehension — Squares

**Explanation:**  
List comprehension is a compact way to write a for loop that builds a list. `[expr for item in iterable]` is equivalent to a for loop appending `expr` each time. It's Pythonic and efficient.

````python
squares = [x**2 for x in range(1, 11)]
print("Squares:", squares)
````

## ✅ Program 9: Iterate Dictionary with `items()`

**Explanation:**  
`dict.items()` returns key-value pairs as tuples. The for loop unpacks each tuple into `key` and `value` variables. This is the standard way to iterate over dictionaries in Python.

````python
student = {"name": "Alice", "age": 20, "grade": "A"}
for key, value in student.items():
    print(f"{key}: {value}")
````

## ✅ Program 10: `enumerate()` — Index + Value

**Explanation:**  
`enumerate()` wraps an iterable and returns `(index, value)` pairs. This avoids manual index tracking and is the Pythonic alternative to `for i in range(len(lst))`.

````python
fruits = ["apple", "banana", "cherry"]
for idx, fruit in enumerate(fruits, start=1):
    print(f"{idx}. {fruit}")
````

## ✅ Program 11: `zip()` — Iterate Two Lists Together

**Explanation:**  
`zip()` pairs elements from two (or more) iterables by position. The loop unpacks each pair into separate variables. It stops at the shorter iterable, making it safe for mismatched lengths.

````python
names = ["Alice", "Bob", "Charlie"]
scores = [85, 92, 78]
for name, score in zip(names, scores):
    print(f"{name}: {score}")
````

## ✅ Program 12: Find Common Elements in Two Lists

**Explanation:**  
We iterate through the first list and check if each element exists in the second list. Elements found in both are added to `common`. The `in` operator makes membership testing readable and concise.

````python
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
common = []
for item in list1:
    if item in list2:
        common.append(item)
print("Common elements:", common)
````

## ✅ Program 13: Remove Duplicates from a List

**Explanation:**  
We build a new list, adding an element only if it hasn't been added already (not in `unique`). This preserves the original order, unlike converting to a set which may reorder elements.

````python
lst = [1, 2, 2, 3, 4, 4, 5, 1]
unique = []
for item in lst:
    if item not in unique:
        unique.append(item)
print("Unique elements:", unique)
````

## ✅ Program 14: Matrix Multiplication

**Explanation:**  
Triple nested for loops compute matrix multiplication. The outer two loops pick the result cell `[i][j]`, and the inner loop accumulates the dot product of row `i` of A and column `j` of B.

````python
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
result = [[0, 0], [0, 0]]
for i in range(2):
    for j in range(2):
        for k in range(2):
            result[i][j] += A[i][k] * B[k][j]
print("Matrix Product:")
for row in result:
    print(row)
````

## ✅ Program 15: Caesar Cipher (Encryption)

**Explanation:**  
Caesar Cipher shifts each letter by a fixed amount (here, 3). We use `ord()` to get ASCII value, shift it, wrap with `% 26`, and convert back with `chr()`. Non-letters are left unchanged.

````python
text = "Hello World"
shift = 3
encrypted = ""
for ch in text:
    if ch.isalpha():
        base = ord('A') if ch.isupper() else ord('a')
        encrypted += chr((ord(ch) - base + shift) % 26 + base)
    else:
        encrypted += ch
print(f"Encrypted: {encrypted}")
````

## ✅ Program 16: Count Words in a Sentence

**Explanation:**  
`split()` breaks a string into a list of words by whitespace. We then iterate over this list and count elements. This is the standard approach to tokenizing and counting words in NLP basics.

````python
sentence = "Python is a powerful and versatile programming language"
word_count = 0
for word in sentence.split():
    word_count += 1
print(f"Word count = {word_count}")
````

## ✅ Program 17: Generate Multiplication Table (Formatted)

**Explanation:**  
Nested for loops generate a formatted multiplication table. The outer loop iterates rows (1–5), the inner loop columns (1–5). `f-string` formatting with fixed width keeps alignment neat.

````python
print("Multiplication Table (1–5):")
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i*j:4}", end="")
    print()
````

## ✅ Program 18: FizzBuzz

**Explanation:**  
FizzBuzz is a classic programming exercise. Print 'Fizz' for multiples of 3, 'Buzz' for multiples of 5, 'FizzBuzz' for both, else the number. Order of conditions matters — check `% 15` first to catch both.

````python
for i in range(1, 31):
    if i % 15 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
````

## ✅ Program 19: Binary Search

**Explanation:**  
Binary search repeatedly halves the search space by comparing the middle element with the target. We use a for loop with an early `break`. This runs in O(log n) compared to O(n) for linear search.

````python
lst = [2, 5, 8, 12, 16, 23, 38, 45]
target = 23
low, high = 0, len(lst) - 1
found = False
for _ in range(len(lst)):
    mid = (low + high) // 2
    if lst[mid] == target:
        print(f"Found {target} at index {mid}")
        found = True
        break
    elif lst[mid] < target:
        low = mid + 1
    else:
        high = mid - 1
if not found:
    print(f"{target} not found")
````

## ✅ Program 20: Bubble Sort

**Explanation:**  
Bubble Sort repeatedly compares adjacent elements and swaps them if out of order. After each pass, the largest unsorted element "bubbles" to the end. Nested for loops perform `n-1` passes with reducing comparisons.

````python
arr = [64, 34, 25, 12, 22, 11, 90]
n = len(arr)
for i in range(n - 1):
    for j in range(n - 1 - i):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]
print("Sorted array:", arr)
````

# 🔷 PART 3: Pattern Printing Programs
<a id='patterns'></a>

> Pattern printing uses **nested loops**: the **outer loop** controls the **number of rows** and the **inner loop** controls the **characters per row**.
>
> Key: `print()` adds a newline at end of each row; `end=" "` suppresses it.

---

## 🔷 Pattern 1: Right-Angled Triangle of Stars

**Explanation:**  
Row `i` contains `i` stars. The inner loop runs exactly `i` times for each row `i`. This produces a right-angled triangle aligned to the left.

````python
n = 5
for i in range(1, n + 1):
    for j in range(i):
        print("*", end=" ")
    print()
````

## 🔷 Pattern 2: Inverted Right-Angled Triangle

**Explanation:**  
Row `i` contains `n - i + 1` stars, decreasing as `i` increases. This creates a descending staircase pattern — the mirror of Pattern 1.

````python
n = 5
for i in range(n, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
````

## 🔷 Pattern 3: Full Pyramid of Stars

**Explanation:**  
A pyramid has leading spaces to center the stars. Row `i` has `n - i` spaces and `2*i - 1` stars. The combination of spaces and stars creates the triangular pyramid shape.

````python
n = 5
for i in range(1, n + 1):
    print(" " * (n - i) + "* " * i)
````

## 🔷 Pattern 4: Inverted Pyramid

**Explanation:**  
The inverted pyramid starts with the most stars and reduces each row. Spaces increase as rows progress downward. This is the flipped version of Pattern 3.

````python
n = 5
for i in range(n, 0, -1):
    print(" " * (n - i) + "* " * i)
````

## 🔷 Pattern 5: Diamond Pattern

**Explanation:**  
A diamond is a pyramid on top of an inverted pyramid. We print the upper half (rows 1 to n) and the lower half (rows n-1 to 1). The widest row at `i = n` is the middle of the diamond.

````python
n = 5
for i in range(1, n + 1):
    print(" " * (n - i) + "* " * i)
for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "* " * i)
````

## 🔷 Pattern 6: Number Triangle

**Explanation:**  
Instead of stars, we print numbers. Row `i` contains numbers from 1 to `i`. This shows that inner loop variables can directly serve as the printed content.

````python
n = 5
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
````

## 🔷 Pattern 7: Pascal's Triangle

**Explanation:**  
Pascal's Triangle uses binomial coefficients. Each row starts and ends with 1; inner values are `prev[j-1] + prev[j]`. We carry forward the previous row to compute the next.

````python
n = 6
row = [1]
for i in range(n):
    print(" " * (n - i) + " ".join(str(x) for x in row))
    row = [1] + [row[j] + row[j+1] for j in range(len(row)-1)] + [1]
````

## 🔷 Pattern 8: Hollow Rectangle

**Explanation:**  
A hollow rectangle prints `*` only on the border (first/last row or first/last column) and spaces inside. The condition `i in (1, rows) or j in (1, cols)` identifies border positions.

````python
rows, cols = 5, 10
for i in range(1, rows + 1):
    for j in range(1, cols + 1):
        if i in (1, rows) or j in (1, cols):
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

## 🔷 Pattern 9: Floyd's Triangle

**Explanation:**  
Floyd's Triangle fills rows with consecutive natural numbers. We maintain a running counter `num` that increments with each print. Row `i` contains `i` numbers.

````python
n = 5
num = 1
for i in range(1, n + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
````

## 🔷 Pattern 10: Alphabet Triangle

**Explanation:**  
`chr(65 + j)` converts integers to uppercase letters (A=65). Row `i` prints the first `i` letters of the alphabet. This combines ASCII arithmetic with loop control.

````python
n = 5
for i in range(1, n + 1):
    for j in range(i):
        print(chr(65 + j), end=" ")
    print()
````

## 🔷 Pattern 11: Zigzag / Butterfly Pattern

**Explanation:**  
The butterfly pattern has stars on both sides expanding outward then inward. Stars on the left: `i` stars; stars on the right: `i` stars; spaces between: `2 * (n - i)`. Two loops produce top and bottom halves.

````python
n = 5
for i in range(1, n + 1):
    print("*" * i + " " * (2 * (n - i)) + "*" * i)
for i in range(n, 0, -1):
    print("*" * i + " " * (2 * (n - i)) + "*" * i)
````

## 🔷 Pattern 12: Checkerboard Pattern

**Explanation:**  
A checkerboard alternates `*` and space based on whether `i + j` is even or odd. This simulates a chess-like grid using simple arithmetic on loop indices.

````python
rows, cols = 6, 6
for i in range(rows):
    for j in range(cols):
        if (i + j) % 2 == 0:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

## 🔷 Pattern 13: Right-Aligned Triangle

**Explanation:**  
Leading spaces push stars to the right. Row `i` has `n - i` spaces followed by `i` stars. This creates the mirror image of Pattern 1, aligned to the right.

````python
n = 5
for i in range(1, n + 1):
    print(" " * (n - i) + "* " * i)
````

## 🔷 Pattern 14: Hourglass Pattern

**Explanation:**  
An hourglass is an inverted pyramid followed by a regular pyramid. Stars decrease to 1 in the middle, then increase again. Spaces are added to center each row.

````python
n = 5
for i in range(n, 0, -1):
    print(" " * (n - i) + "* " * i)
for i in range(2, n + 1):
    print(" " * (n - i) + "* " * i)
````

## 🔷 Pattern 15: Solid Square

**Explanation:**  
The simplest 2D pattern: `n × n` grid of stars. Outer loop = rows, inner loop = columns. Every cell is filled — no conditions needed.

````python
n = 5
for i in range(n):
    for j in range(n):
        print("*", end=" ")
    print()
````

## 🔷 Pattern 16: X Pattern

**Explanation:**  
Stars appear on the two diagonals: main diagonal where `i == j` and anti-diagonal where `i + j == n - 1`. All other positions print a space. Together they form an X shape.

````python
n = 7
for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

## 🔷 Pattern 17: Spiral Number Matrix (Concept)

**Explanation:**  
We simulate a simple 2D matrix filled row-by-row with sequential numbers. Though not a true spiral, this demonstrates how to populate and display a 2D grid — a foundation for more complex spiral algorithms.

````python
n = 4
matrix = [[0]*n for _ in range(n)]
num = 1
for i in range(n):
    for j in range(n):
        matrix[i][j] = num
        num += 1

print("Number Matrix:")
for row in matrix:
    print("".join(f"{x:4}" for x in row))
````

## 🔷 Pattern 18: Number Pyramid (Same Number Each Row)

**Explanation:**  
Each row prints the row number `i` repeated `i` times. Unlike Pattern 6 (sequential numbers), here every position in row `i` contains the value `i`. This creates a staircase of repeated digits.

````python
n = 5
for i in range(1, n + 1):
    for j in range(i):
        print(i, end=" ")
    print()
````

## 🔷 Pattern 19: Star Border with Hollow Inside

**Explanation:**  
This prints a hollow diamond shape. Stars appear only on the outer edges of each row, with spaces filling the interior. The condition checks if `j` is the first or last position in that row's range.

````python
n = 5
# Upper half
for i in range(1, n + 1):
    spaces = " " * (n - i)
    if i == 1:
        print(spaces + "*")
    else:
        print(spaces + "*" + " " * (2 * i - 3) + "*")
# Lower half
for i in range(n - 1, 0, -1):
    spaces = " " * (n - i)
    if i == 1:
        print(spaces + "*")
    else:
        print(spaces + "*" + " " * (2 * i - 3) + "*")
````

## 🔷 Pattern 20: Alternating 0-1 Triangle

**Explanation:**  
Each position `(i, j)` prints 1 if `(i + j) % 2 == 0` (or if the row starts with 1), else 0. The starting value alternates between rows, creating a checkerboard-style triangle pattern.

````python
n = 5
for i in range(1, n + 1):
    start = 1 if i % 2 != 0 else 0
    for j in range(i):
        print((start + j) % 2, end=" ")
    print()
````

## 📌 Summary Table

| Feature | `while` loop | `for` loop |
|--------|-------------|------------|
| **Used when** | Condition-based iteration | Known number of iterations |
| **Iteration control** | Manual (increment/decrement) | Automatic via sequence |
| **Risk** | Infinite loop if condition never false | Safer by default |
| **Best for** | Input validation, game loops, sensors | Lists, strings, ranges, files |
| **`else` clause** | Runs if condition becomes False | Runs if no `break` occurred |

---

> 💡 **Pro Tip:** For most tasks involving sequences, prefer `for` loops. Reserve `while` loops for situations where termination depends on a condition that isn't tied to a fixed count.
