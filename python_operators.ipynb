# 🐍 Python Operators — Complete Guide

> **Operators** are special symbols that tell Python to perform specific operations on values (called **operands**).
> Think of them as the **verbs** of Python — they make things *happen*.

**Real Life Analogy:** A calculator's `+`, `-`, `×`, `÷` buttons are operators. Python has all of those — and many more!

## 1️⃣ Arithmetic Operators

**Used to perform basic mathematical calculations**

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `10 + 3` | `13` |
| `-` | Subtraction | `10 - 3` | `7` |
| `*` | Multiplication | `10 * 3` | `30` |
| `/` | Division (float) | `10 / 3` | `3.333...` |
| `//` | Floor Division | `10 // 3` | `3` |
| `%` | Modulus (remainder) | `10 % 3` | `1` |
| `**` | Exponentiation | `10 ** 3` | `1000` |

> **Key difference:** `/` always returns a float. `//` returns an integer (rounds down).

**Real Life Uses:**
- `+` : Adding items in a cart
- `%` : Checking if a number is even/odd, finding leap years
- `**` : Calculating compound interest, areas, powers
- `//` : Finding how many full groups can be formed


````python
# ── Basic Arithmetic ──
a = 17
b = 5

print('--- Basic Operations ---')
print(f'{a} + {b}  = {a + b}')   # Addition
print(f'{a} - {b}  = {a - b}')   # Subtraction
print(f'{a} * {b}  = {a * b}')   # Multiplication
print(f'{a} / {b}  = {a / b}')   # True Division -> float
print(f'{a} // {b} = {a // b}')  # Floor Division -> integer
print(f'{a} % {b}  = {a % b}')   # Modulus -> remainder
print(f'{a} ** {b} = {a ** b}')  # Exponent

print()
````

## 2️⃣ Comparison (Relational) Operators

**Used to compare two values — always returns `True` or `False`**

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `7 > 3` | `True` |
| `<` | Less than | `3 < 7` | `True` |
| `>=` | Greater than or equal | `5 >= 5` | `True` |
| `<=` | Less than or equal | `4 <= 5` | `True` |

> **Important:** `==` checks **equality** (two equal signs). `=` is **assignment** (one equal sign). Never mix them up!

**Real Life Uses:**
- `==` : Login password verification
- `>` : Checking if temperature exceeds limit
- `<=` : Age limit check for a platform
- `!=` : Flagging mismatched data


````python
# ── Basic Comparisons ──
x, y = 10, 20
print('--- Basic Comparisons ---')
print(f'{x} == {y}  -> {x == y}')   # Equal
print(f'{x} != {y}  -> {x != y}')   # Not equal
print(f'{x} >  {y}  -> {x > y}')    # Greater than
print(f'{x} <  {y}  -> {x < y}')    # Less than
print(f'{x} >= {y}  -> {x >= y}')   # Greater or equal
print(f'{x} <= {y}  -> {x <= y}')   # Less or equal

print()

# ── Real Example 1: Password check ──
stored_password = 'python@123'
entered_password = 'python@123'
if entered_password == stored_password:
    print('🔓 Login Successful!')
else:
    print('🔒 Wrong Password. Try again.')

# ── Real Example 2: Temperature alert system ──
temperatures = [34, 38, 41, 29, 45]
print()
for temp in temperatures:
    if temp >= 40:
        print(f'🌡️  {temp}°C — HEAT ALERT! Stay indoors.')
    elif temp >= 35:
        print(f'🌤️  {temp}°C — Hot day. Stay hydrated.')
    else:
        print(f'😊  {temp}°C — Pleasant weather.')

# ── Real Example 3: Exam pass/fail with distinction ──
print()
scores = [92, 75, 49, 60, 85]
for score in scores:
    if score >= 90:
        print(f'Score {score}: 🏆 Distinction')
    elif score >= 60:
        print(f'Score {score}: ✅ Pass')
    else:
        print(f'Score {score}: ❌ Fail')

# ── Real Example 4: Stock price alert ──
print()
buy_price = 1500
current_price = 1750
if current_price > buy_price:
    profit = current_price - buy_price
    print(f'📈 Stock up! Profit of ₹{profit} per share.')
elif current_price < buy_price:
    loss = buy_price - current_price
    print(f'📉 Stock down. Loss of ₹{loss} per share.')
else:
    print('📊 Stock at break-even.')

# ── Real Example 5: Comparing strings ──
print()
city1 = 'Hyderabad'
city2 = 'Chennai'
print(f'"{city1}" == "{city2}" -> {city1 == city2}')
print(f'"{city1}" != "{city2}" -> {city1 != city2}')
````

## 3️⃣ Assignment Operators

**Used to assign values to variables — shorthand for updating a variable**

| Operator | Equivalent to | Example | Result |
|----------|---------------|---------|--------|
| `=` | — | `x = 10` | `x = 10` |
| `+=` | `x = x + n` | `x += 3` | `x = 13` |
| `-=` | `x = x - n` | `x -= 3` | `x = 7` |
| `*=` | `x = x * n` | `x *= 3` | `x = 30` |
| `/=` | `x = x / n` | `x /= 3` | `x = 3.33` |
| `//=` | `x = x // n` | `x //= 3` | `x = 3` |
| `%=` | `x = x % n` | `x %= 3` | `x = 1` |
| `**=` | `x = x ** n` | `x **= 3` | `x = 1000` |

> Compound assignment operators (`+=`, `-=`, etc.) are **shortcuts** — they modify the variable **in place**.

**Real Life Uses:**
- `+=` : Increasing a score, adding items to a total
- `-=` : Deducting from a wallet balance
- `*=` : Scaling a value (e.g., apply a multiplier)


````python
# ── Basic Assignment ──
x = 10
print(f'Initial x = {x}')

x += 5;  print(f'After x += 5  : {x}')   # 15
x -= 3;  print(f'After x -= 3  : {x}')   # 12
x *= 2;  print(f'After x *= 2  : {x}')   # 24
x /= 4;  print(f'After x /= 4  : {x}')   # 6.0
x //= 2; print(f'After x //= 2 : {x}')   # 3.0
x **= 3; print(f'After x **= 3 : {x}')   # 27.0
x %= 5;  print(f'After x %%= 5  : {x}')  # 2.0

print()

# ── Real Example 1: Wallet / UPI balance deductions ──
wallet = 2000.0
print(f'Starting Wallet Balance: ₹{wallet}')
wallet -= 450   # Grocery
print(f'After grocery (₹450)   : ₹{wallet}')
wallet -= 120   # Auto ride
print(f'After auto ride (₹120) : ₹{wallet}')
wallet -= 999   # Online order
print(f'After online order(₹999): ₹{wallet}')

print()

# ── Real Example 2: Game score tracker ──
score = 0
print('🎮 Game Score Tracker')
score += 10; print(f'Collected coin     : Score = {score}')
score += 50; print(f'Defeated enemy     : Score = {score}')
score -= 20; print(f'Hit by obstacle    : Score = {score}')
score += 100; print(f'Completed level!   : Score = {score}')
score *= 2;  print(f'Bonus x2 activated : Score = {score}')

print()

# ── Real Example 3: Countdown timer ──
time_left = 10
print('⏳ Countdown:')
while time_left > 0:
    print(f'  {time_left}...', end=' ')
    time_left -= 3
print('\n🚀 Launch!')

# ── Real Example 4: Compound interest growth using **= ──
print()
amount = 100
amount **= 2   # squared
print(f'100 squared = {amount}')
````

## 4️⃣ Logical Operators

**Used to combine or invert boolean conditions**

| Operator | Description | Example | Result |
|----------|-------------|---------|--------|
| `and` | True if **both** conditions are True | `True and False` | `False` |
| `or` | True if **at least one** condition is True | `True or False` | `True` |
| `not` | **Inverts** the boolean value | `not True` | `False` |

**Full Truth Table:**
```
A       B       A and B    A or B    not A
True    True    True       True      False
True    False   False      True      False
False   True    False      True      True
False   False   False      False     True
```

> **Short-circuit evaluation:**
> - `and` stops at the **first False** (no need to check the rest)
> - `or` stops at the **first True** (no need to check the rest)

**Real Life Uses:**
- `and` : Loan approval needs BOTH good salary AND credit score
- `or` : Free shipping if premium member OR order above ₹500
- `not` : Show login page if NOT logged in


````python
# ── Basic Logical Operations ──
print('--- and ---')
print(f'True  and True  -> {True and True}')
print(f'True  and False -> {True and False}')
print(f'False and False -> {False and False}')

print('--- or ---')
print(f'True  or True   -> {True or True}')
print(f'True  or False  -> {True or False}')
print(f'False or False  -> {False or False}')

print('--- not ---')
print(f'not True  -> {not True}')
print(f'not False -> {not False}')

print()

# ── Real Example 1: Loan eligibility ──
salary = 45000
credit_score = 720
employment = 'permanent'
if salary >= 30000 and credit_score >= 700 and employment == 'permanent':
    print('✅ Loan Approved — all criteria met!')
else:
    print('❌ Loan Rejected — one or more criteria failed.')

# ── Real Example 2: Free delivery (OR logic) ──
print()
is_premium = False
order_amount = 650
has_coupon = True
if is_premium or order_amount >= 500 or has_coupon:
    print('🚚 Free Delivery Applied!')
else:
    print('📦 Delivery fee: ₹49')

# ── Real Example 3: not — redirect to login ──
print()
is_logged_in = False
if not is_logged_in:
    print('🔐 You are not logged in. Redirecting to login page...')

# ── Real Example 4: Combining all three ──
print()
age = 22
has_id = True
is_blacklisted = False
if age >= 18 and has_id and not is_blacklisted:
    print('🎉 Entry allowed to the venue!')
else:
    print('🚫 Entry denied.')

# ── Real Example 5: Flight booking — seat class ──
print()
miles = 12000
is_elite = True
if miles >= 10000 or is_elite:
    print('✈️  Upgrade to Business Class available!')
else:
    print('🪑 Economy class only.')

# ── Real Example 6: Short-circuit demo ──
print()
x = 0
# Division never happens because x == 0 is False, and stops early
if x != 0 and (100 / x > 5):
    print('Condition met')
else:
    print(f'Short-circuit: x={x}, second condition never evaluated (safe from ZeroDivisionError)')
````

## 5️⃣ Bitwise Operators

**Operate directly on the binary (bit-level) representation of integers**

| Operator | Name | Description |
|----------|------|-------------|
| `&` | AND | 1 if **both** bits are 1 |
| `\|` | OR | 1 if **either** bit is 1 |
| `^` | XOR | 1 if bits are **different** |
| `~` | NOT | **Flips** all bits (`~n = -(n+1)`) |
| `<<` | Left Shift | Shifts bits left (multiplies by 2) |
| `>>` | Right Shift | Shifts bits right (divides by 2) |

**Binary Example: 12 and 10**
```
12 in binary = 1100
10 in binary = 1010

12 & 10  ->  1100 AND 1010 = 1000 = 8
12 | 10  ->  1100  OR 1010 = 1110 = 14
12 ^ 10  ->  1100 XOR 1010 = 0110 = 6
~12      ->  -(12+1)       =      = -13
12 << 1  ->  11000         =      = 24  (x2)
12 >> 1  ->  0110          =      = 6   (/2)
```

**Real Life Uses:**
- Checking even/odd with `&` faster than `%`
- Permission flags in operating systems
- Efficient multiply/divide by powers of 2
- Cryptography and data encryption


````python
# ── Binary representation helper ──
def show_bits(label, value):
    print(f'{label:20s} = {value:4d}  (binary: {bin(value)})')

a = 12   # binary: 1100
b = 10   # binary: 1010

print('--- Bitwise Operations ---')
show_bits('a', a)
show_bits('b', b)
print()
show_bits('a & b  (AND)',  a & b)   # 8
show_bits('a | b  (OR)',   a | b)   # 14
show_bits('a ^ b  (XOR)',  a ^ b)   # 6
print(f'{"~a (NOT)":20s} = {~a}  (binary: {bin(~a)})')  # -13
show_bits('a << 1 (Left)', a << 1)  # 24  (12 * 2)
show_bits('a >> 1 (Right)',a >> 1)  # 6   (12 / 2)

print()

# ── Real Example 1: Fast even/odd check using & ──
print('--- Even/Odd using & ---')
for n in [3, 8, 15, 20, 7]:
    result = 'Odd' if (n & 1) else 'Even'
    print(f'{n:3d} & 1 = {n & 1}  -> {result}')

print()

# ── Real Example 2: Permission flags (like Linux file permissions) ──
READ    = 0b100  # 4
WRITE   = 0b010  # 2
EXECUTE = 0b001  # 1

# Grant READ + WRITE permission
user_permission = READ | WRITE
print(f'User permission bits: {bin(user_permission)} = {user_permission}')
print(f'Can READ?    {bool(user_permission & READ)}')
print(f'Can WRITE?   {bool(user_permission & WRITE)}')
print(f'Can EXECUTE? {bool(user_permission & EXECUTE)}')

print()

# ── Real Example 3: Fast multiply/divide by 2 using shifts ──
print('--- Left/Right Shift ---')
val = 8
print(f'{val} << 1 = {val << 1}  (x2)')
print(f'{val} << 2 = {val << 2}  (x4)')
print(f'{val} >> 1 = {val >> 1}  (/2)')
print(f'{val} >> 2 = {val >> 2}  (/4)')

print()

# ── Real Example 4: Toggle a bit with XOR ──
flag = 0b1010
toggle = 0b0010  # flip the second bit
print(f'Original   : {bin(flag)}')
print(f'After XOR  : {bin(flag ^ toggle)}')
print(f'After again: {bin(flag ^ toggle ^ toggle)}  (toggled back)')
````

## 6️⃣ Membership Operators

**Check whether a value exists inside a sequence (string, list, tuple, set, dict)**

| Operator | Description | Example | Result |
|----------|-------------|---------|--------|
| `in` | True if value **is found** in the sequence | `'a' in 'cat'` | `True` |
| `not in` | True if value is **not found** in the sequence | `5 not in [1, 2, 3]` | `True` |

**Works with:**
- `str` — checks for substrings
- `list`, `tuple` — checks for elements
- `set` — fastest membership check (O(1))
- `dict` — checks for **keys** by default

**Real Life Uses:**
- Checking if a username is taken
- Validating if a product is in stock
- Email spam filter — checking blacklisted words
- Checking if a city is in a list of serviceable areas


````python
# ── With Strings ──
print('--- Membership in Strings ---')
sentence = 'Python is awesome for data science'
print(f'"Python" in sentence     -> {"Python" in sentence}')
print(f'"Java" in sentence       -> {"Java" in sentence}')
print(f'"Java" not in sentence   -> {"Java" not in sentence}')
print(f'"data" in sentence       -> {"data" in sentence}')

print()

# ── With Lists ──
print('--- Membership in Lists ---')
fruits = ['apple', 'banana', 'mango', 'cherry', 'grape']
for item in ['mango', 'pineapple', 'banana', 'watermelon']:
    if item in fruits:
        print(f'  ✅ "{item}" is available')
    else:
        print(f'  ❌ "{item}" is NOT in stock')

print()

# ── Real Example 1: Serviceable cities (e-commerce) ──
serviceable_cities = ['Hyderabad', 'Bengaluru', 'Mumbai', 'Chennai', 'Pune', 'Delhi']
customer_city = 'Amaravati'
if customer_city in serviceable_cities:
    print(f'📦 Delivery available in {customer_city}!')
else:
    print(f'🚫 Sorry, we do not deliver to {customer_city} yet.')

print()

# ── Real Example 2: Spam filter using membership ──
spam_words = ['win', 'free', 'prize', 'click', 'urgent', 'offer']
emails = [
    'You have won a free prize click here!',
    'Meeting scheduled for tomorrow at 3pm',
    'Urgent! Claim your exclusive offer now',
    'Please review the attached report'
]
print('--- Spam Filter ---')
for email in emails:
    lower_email = email.lower()
    is_spam = any(word in lower_email for word in spam_words)
    status = '🚨 SPAM' if is_spam else '✅ INBOX'
    print(f'  [{status}] {email[:40]}...')

print()

# ── Real Example 3: Membership in dict (checks keys) ──
student_scores = {'Arjun': 88, 'Priya': 92, 'Ravi': 74}
for name in ['Priya', 'Sneha', 'Arjun']:
    if name in student_scores:
        print(f'  {name}: Score = {student_scores[name]}')
    else:
        print(f'  {name}: No record found')

print()

# ── Real Example 4: Vowel checker ──
vowels = 'aeiouAEIOU'
test_word = 'Programming'
print(f'Vowels in "{test_word}":')
for ch in test_word:
    if ch in vowels:
        print(f'  "{ch}" is a vowel')
````

## 7️⃣ Identity Operators

**Check whether two variables point to the exact same object in memory**

| Operator | Description | Example | Result |
|----------|-------------|---------|--------|
| `is` | True if both refer to the **same object** | `a is b` | depends |
| `is not` | True if they refer to **different objects** | `a is not b` | depends |

> **`==` vs `is` — the most common confusion in Python!**
> - `==` checks if values are **equal**
> - `is` checks if they are the **same object in memory** (same `id()`)

```
a = [1, 2, 3]
b = [1, 2, 3]
a == b   -> True   (values are equal)
a is b   -> False  (different objects in memory)
```

**Python caches small integers (-5 to 256) and short strings — so `is` may return `True` for them.**

**Real Life Uses:**
- Checking if a variable is `None` (always use `is None`, not `== None`)
- Verifying singleton objects
- Understanding Python's object model


````python
# ── Basic Identity ──
a = [1, 2, 3]
b = [1, 2, 3]   # same values, different object
c = a           # same object (c points to a)

print('--- Identity vs Equality ---')
print(f'a = {a},  id(a) = {id(a)}')
print(f'b = {b},  id(b) = {id(b)}')
print(f'c = {c},  id(c) = {id(c)}')
print()
print(f'a == b    -> {a == b}   (same values?)')
print(f'a is b    -> {a is b}  (same object?)')
print(f'a is c    -> {a is c}   (same object?)')
print(f'a is not b -> {a is not b}  (different objects?)')

print()

# ── Real Example 1: None check (most common use of `is`) ──
def find_user(name):
    users = {'Arjun': 101, 'Priya': 102}
    return users.get(name)   # returns None if not found

for name in ['Arjun', 'Ravi']:
    result = find_user(name)
    if result is None:
        print(f'  User "{name}" not found.')
    else:
        print(f'  User "{name}" found! ID = {result}')

print()

# ── Real Example 2: Small integer caching (Python internals) ──
x = 100
y = 100
print(f'x = y = 100')
print(f'x == y  -> {x == y}')
print(f'x is y  -> {x is y}  (cached! same object for small ints)')

print()

p = 1000
q = 1000
print(f'p = q = 1000')
print(f'p == q  -> {p == q}')
print(f'p is q  -> {p is q}  (not cached, different objects for large ints)')

print()

# ── Real Example 3: Checking type using is ──
values = [42, 'hello', None, 3.14, True]
print('--- Type identity check ---')
for v in values:
    if v is None:
        print(f'  {repr(v):10s} -> is None')
    elif type(v) is int:
        print(f'  {repr(v):10s} -> is int')
    elif type(v) is str:
        print(f'  {repr(v):10s} -> is str')
    else:
        print(f'  {repr(v):10s} -> other type: {type(v).__name__}')

print()
# ── Best Practice ──
result = None
print('Best practice: always use "is None" not "== None"')
print(f'result is None     -> {result is None}')      # correct
print(f'result == None     -> {result == None}')      # works but not recommended
````

## 8️⃣ Operator Precedence

**When multiple operators appear in one expression, Python follows a strict order — just like BODMAS in maths.**

**Precedence Table (Highest to Lowest):**

| Priority | Operator(s) | Description |
|----------|-------------|-------------|
| 1 (Highest) | `()` | Parentheses |
| 2 | `**` | Exponentiation |
| 3 | `~x`, `+x`, `-x` | Unary NOT, Plus, Minus |
| 4 | `*`, `/`, `//`, `%` | Multiplication, Division, Floor, Modulus |
| 5 | `+`, `-` | Addition, Subtraction |
| 6 | `<<`, `>>` | Bitwise Shifts |
| 7 | `&` | Bitwise AND |
| 8 | `^` | Bitwise XOR |
| 9 | `\|` | Bitwise OR |
| 10 | `==`, `!=`, `>`, `<`, `>=`, `<=`, `is`, `is not`, `in`, `not in` | Comparisons / Membership / Identity |
| 11 | `not` | Logical NOT |
| 12 | `and` | Logical AND |
| 13 (Lowest) | `or` | Logical OR |

> **Golden Rule:** When in doubt, use **parentheses `()`** to make the order explicit and your code readable!

**BODMAS reminder:**
```
B - Brackets     ()          <- highest priority
O - Orders       **
D - Division     /  //
M - Multiplication *
A - Addition     +
S - Subtraction  -           <- lower priority
```


````python
# ── Precedence Examples ──
print('--- Precedence Demos ---')

# Arithmetic precedence
print(f'2 + 3 * 4       = {2 + 3 * 4}       (* before +, result: 14)')
print(f'(2 + 3) * 4     = {(2 + 3) * 4}       (parens first, result: 20)')
print(f'2 ** 3 ** 2     = {2 ** 3 ** 2}     (** is right-associative: 2^9=512)')
print(f'(2 ** 3) ** 2   = {(2 ** 3) ** 2}    (left first: 8^2=64)')
print(f'10 - 4 / 2      = {10 - 4 / 2}      (/ before -, result: 8.0)')
print(f'10 - 4 // 2 + 1 = {10 - 4 // 2 + 1}  (// then - then +, result: 9)')

print()

# ── Mixed operators ──
print('--- Mixed Precedence ---')
result1 = 5 + 3 > 2 * 4 - 1
# Step by step: 2*4=8 -> 8-1=7 -> 5+3=8 -> 8>7 -> True
print(f'5 + 3 > 2 * 4 - 1    = {result1}  (8 > 7)')

result2 = not True or False and True
# Step by step: not True = False -> False and True = False -> False or False = False
print(f'not True or False and True = {result2}  (not > and > or)')

result3 = True or False and False
# and evaluated first: False and False = False -> True or False = True
print(f'True or False and False    = {result3}   (and before or)')

print()

# ── Real Example 1: Bill with tax and discount ──
price = 500
tax_rate = 0.18
discount = 50

wrong  = price + price * tax_rate - discount   # common mistake thinking: (price+price)*tax_rate
correct = (price - discount) + (price - discount) * tax_rate
print(f'Price=₹{price}, Tax={int(tax_rate*100)}%, Discount=₹{discount}')
print(f'Without careful precedence : ₹{wrong}')
print(f'With correct parentheses   : ₹{correct:.2f}')

print()

# ── Real Example 2: Access control — precedence with logical ops ──
is_admin = False
is_verified = True
has_permission = True

# Without parens: is_admin or (is_verified and has_permission) <- 'and' binds tighter
result_a = is_admin or is_verified and has_permission
# With parens: (is_admin or is_verified) and has_permission
result_b = (is_admin or is_verified) and has_permission

print(f'is_admin={is_admin}, is_verified={is_verified}, has_permission={has_permission}')
print(f'is_admin or is_verified and has_permission      = {result_a}  (and runs first)')
print(f'(is_admin or is_verified) and has_permission    = {result_b}  (parens override)')

print()

# ── Real Example 3: Step-by-step expression breakdown ──
expr = 3 + 2 ** 2 * 4 - 8 // 3
# Step 1: 2**2 = 4
# Step 2: 4 * 4 = 16
# Step 3: 8 // 3 = 2
# Step 4: 3 + 16 - 2 = 17
print(f'3 + 2 ** 2 * 4 - 8 // 3')
print(f'= 3 + 4 * 4 - 8 // 3   (** first)')
print(f'= 3 + 16 - 2           (* and // next)')
print(f'= {expr}              (+ and - last)')
````

## 📋 Complete Operators Cheat Sheet

```python
# ARITHMETIC
a + b    # Addition
a - b    # Subtraction
a * b    # Multiplication
a / b    # Division (float)
a // b   # Floor Division
a % b    # Modulus
a ** b   # Exponent

# COMPARISON
a == b   # Equal
a != b   # Not equal
a > b    # Greater than
a < b    # Less than
a >= b   # Greater or equal
a <= b   # Less or equal

# ASSIGNMENT
x = 5; x += 2; x -= 2; x *= 2

# LOGICAL
a and b  # Both True
a or b   # At least one True
not a    # Invert

# BITWISE
a & b    # AND
a | b    # OR
a ^ b    # XOR
~a       # NOT
a << n   # Left shift
a >> n   # Right shift

# MEMBERSHIP
x in seq
x not in seq

# IDENTITY
a is b
```

## 💡 Key Takeaways

| Rule | Tip |
|------|-----|
| `==` vs `=` | `==` compares, `=` assigns |
| `==` vs `is` | `==` checks value, `is` checks memory address |
| `None` check | Always write `x is None`, not `x == None` |
| Precedence | Use `()` to make order clear and avoid bugs |
| `and` vs `or` | `and` has higher precedence than `or` |
| `not` | `not` has higher precedence than `and` and `or` |
| Bitwise | Works only on integers |
| `in` with dict | Checks **keys**, not values |
