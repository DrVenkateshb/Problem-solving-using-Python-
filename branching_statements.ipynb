# 🐍 Branching Statements in Python

> **Branching** lets your program make **decisions**.
> Think of it like a crossroad — based on a condition, Python takes one path or another.

**Real Life Analogy:** A security guard at a club checks your age and lets you in or turns you away.
In Python, an `if` statement checks a condition and runs one block or another.

---

## 📌 Topics Covered

| # | Statement | Purpose |
|---|-----------|--------|
| 1 | `if` | Run code only when condition is True |
| 2 | `if-else` | Two outcomes — True path or False path |
| 3 | `if-elif-else` | Multiple conditions checked one by one |
| 4 | Nested `if` | An `if` inside another `if` |
| 5 | Logical Operators | Combine conditions with `and`, `or`, `not` |
| 6 | Ternary Expression | One-line shorthand decision |
| 7 | Mini Project | Real-world electricity bill calculator |


---
## 1️⃣ The `if` Statement

**Do this ONLY if the condition is True**

```
if condition:
    code to run   <- only executes when condition is True
```

- If the condition is **True** -> the indented block runs
- If the condition is **False** -> Python skips the block entirely
- The colon `:` after `if` is **mandatory**
- Indentation (4 spaces) defines the block

**Real Life Example:** Checking if a shop is open before entering.


````python
# Example 1: Store open check
store_open = True

if store_open:
    print('🛒 Come in! The store is open.')

# Example 2: Voting eligibility
age = 20

if age >= 18:
    print('✅ You are eligible to vote!')

print('--- Program continues regardless ---')
````

---
## 2️⃣ The `if-else` Statement

**Do THIS if True, otherwise do THAT**

```
if condition:
    code when True     <- runs if condition is True
else:
    code when False    <- runs if condition is False
```

- `else` has **no condition** — it is the automatic fallback
- Exactly **one** of the two blocks will always run
- Together they cover **every possible outcome**

**Real Life Example:** ATM checks your balance before allowing withdrawal.


````python
# Example 1: ATM withdrawal
balance = 500
withdrawal = 1000

if withdrawal <= balance:
    print('💸 Transaction successful!')
else:
    print('❌ Insufficient balance. Please try a smaller amount.')

# Example 2: Even or Odd
number = 7

if number % 2 == 0:
    print(f'{number} is an Even number 🟢')
else:
    print(f'{number} is an Odd number 🔵')
````

---
## 3️⃣ The `if-elif-else` Statement

**Check MULTIPLE conditions one by one**

```
if condition_1:
    code for condition 1
elif condition_2:        <- checked only when above is False
    code for condition 2
elif condition_3:
    code for condition 3
else:
    fallback code        <- runs when NONE of the above matched
```

- `elif` means **else if** — Python checks top to bottom
- As soon as one condition is True, Python runs that block and **skips the rest**
- You can have **as many `elif`s** as you need
- The final `else` is optional but good practice

**Real Life Example:** Exam grading — your marks fall into exactly one grade band.


````python
# Example 1: Exam grade calculator
marks = 85

if marks >= 90:
    grade = 'A+ 🏆 Outstanding!'
elif marks >= 80:
    grade = 'A  🌟 Excellent!'
elif marks >= 70:
    grade = 'B  👍 Good job!'
elif marks >= 60:
    grade = 'C  📘 Keep it up!'
elif marks >= 50:
    grade = 'D  ⚠️  You passed.'
else:
    grade = 'F  ❌ Better luck next time.'

print(f'Marks: {marks} -> Grade: {grade}')

# Example 2: Traffic light system
light = 'green'

if light == 'green':
    print('🟢 GO — Move forward!')
elif light == 'yellow':
    print('🟡 SLOW DOWN — Prepare to stop!')
elif light == 'red':
    print('🔴 STOP — Wait for green!')
else:
    print('⚠️  Unknown signal — proceed with caution!')
````

## 4️⃣ Nested `if` Statements

**An `if` inside another `if`**

```
if outer_condition:
    # outer block
    if inner_condition:
        # inner block — runs only when BOTH conditions are True
```

- Used when a decision depends on **another decision**
- Each level of nesting adds **4 more spaces** of indentation
- Avoid going deeper than **3 levels** — it gets hard to read

**Real Life Example:** A cinema first checks if you have a ticket, then checks your age for the film rating.


````python
# Example 1: Movie ticket + age check
has_ticket = True
age = 16

if has_ticket:
    print('🎟️  Ticket found!')
    if age >= 18:
        print('🎬 Enjoy the adult movie!')
    else:
        print('🧒 Only family-friendly movies for you.')
else:
    print('🚫 No ticket — please buy one first.')

# Example 2: Online shopping discount
is_member = True
cart_total = 1200

if is_member:
    if cart_total >= 1000:
        discount = 20
        print(f'🎉 Premium member discount: {discount}% off!')
    else:
        discount = 10
        print(f'✅ Member discount: {discount}% off!')
else:
    if cart_total >= 2000:
        discount = 5
        print(f'👋 Guest discount: {discount}% off for large orders!')
    else:
        discount = 0
        print('💡 Sign up as a member to get discounts!')

final_price = cart_total - (cart_total * discount / 100)
print(f'💰 Final Price: ₹{final_price}')
````

## 5️⃣ Logical Operators in Conditions

**Combine or invert conditions using `and`, `or`, `not`**

| Operator | Meaning | Result is True when... |
|----------|---------|------------------------|
| `and` | Both must be True | **All** conditions are True |
| `or` | At least one True | **Any** condition is True |
| `not` | Flips the result | The condition is **False** |

**Truth Table:**
```
True  and True  -> True
True  and False -> False
True  or  False -> True
False or  False -> False
not True        -> False
not False       -> True
```

**Real Life Example:** A bank checks BOTH your salary AND credit score before approving a loan.


````python
# Example 1: Loan eligibility (AND — both must pass)
salary = 50000
credit_score = 750

if salary >= 30000 and credit_score >= 700:
    print('✅ Loan Approved! You meet all criteria.')
else:
    print('❌ Loan Denied. Please improve your eligibility.')

# Example 2: Free delivery (OR — either condition is enough)
is_premium = False
order_value = 600

if is_premium or order_value >= 500:
    print('🚚 Free delivery on your order!')
else:
    print('📦 Delivery charge: ₹50')

# Example 3: Login check (NOT — flip the condition)
is_logged_in = False

if not is_logged_in:
    print('🔐 Please log in to continue.')
````

## 6️⃣ Ternary Expression (One-Line `if`)

**A compact shorthand for simple True/False decisions**

```python
value = result_if_true  if  condition  else  result_if_false
#        ^ what you get      ^ check here     ^ what you get otherwise
```

- Same logic as `if-else`, written in **one line**
- Best used for **simple assignments** — not complex logic
- Makes your code **shorter and more readable** when appropriate

**Real Life Example:** A price tag that reads 'On Sale' or 'Regular Price' based on the price.


````python
# Example 1: Pass or Fail
score = 55
result = 'Pass ✅' if score >= 50 else 'Fail ❌'
print(f'Result: {result}')

# Example 2: Discount label
price = 999
label = '💰 On Sale!' if price < 1000 else 'Regular Price'
print(label)

# Example 3: Greeting based on time of day
hour = 9
greeting = 'Good Morning 🌅' if hour < 12 else 'Good Afternoon 🌞'
print(greeting)

# Note: the ternary above is equivalent to:
# if score >= 50:
#     result = 'Pass ✅'
# else:
#     result = 'Fail ❌'
````

## 7️⃣ Real-World Mini Project — Electricity Bill Calculator ⚡

**Problem:** Calculate an electricity bill based on units consumed using slab-based pricing.

| Units Consumed | Rate per Unit |
|---------------|---------------|
| 0 – 100 | ₹3.50 |
| 101 – 200 | ₹5.00 |
| 201 – 400 | ₹7.00 |
| Above 400 | ₹9.00 |

> Each slab's units are charged at that slab's rate only — not the full amount at the highest rate.
> This is exactly how your actual electricity board calculates your monthly bill!


````python
# ⚡ Electricity Bill Calculator
# Change the value of 'units' below and re-run to test!

units = 320

if units <= 100:
    bill = units * 3.50
    category = 'Low Usage 🌱'
elif units <= 200:
    bill = (100 * 3.50) + ((units - 100) * 5.00)
    category = 'Moderate Usage 💡'
elif units <= 400:
    bill = (100 * 3.50) + (100 * 5.00) + ((units - 200) * 7.00)
    category = 'High Usage ⚡'
else:
    bill = (100 * 3.50) + (100 * 5.00) + (200 * 7.00) + ((units - 400) * 9.00)
    category = 'Very High Usage 🔥'

print('=' * 42)
print('     ⚡  ELECTRICITY BILL SUMMARY  ⚡')
print('=' * 42)
print(f'  Units Consumed  :  {units} kWh')
print(f'  Category        :  {category}')
print(f'  Total Bill      :  ₹{bill:.2f}')
print('=' * 42)
````

## 📋 Quick Reference Cheat Sheet

```python
# 1. if — one condition
if condition:
    ...

# 2. if-else — two outcomes
if condition:
    ...
else:
    ...

# 3. if-elif-else — multiple conditions
if condition_1:
    ...
elif condition_2:
    ...
else:
    ...

# 4. Nested if
if outer:
    if inner:
        ...

# 5. Logical operators
if a and b:   ...   # both True
if a or b:    ...   # at least one True
if not a:     ...   # a is False

# 6. Ternary
x = 'yes' if condition else 'no'
```

---

## 💡 Golden Rules to Remember

1. Always use a colon `:` after every `if`, `elif`, and `else`
2. **Indentation is mandatory** — use exactly 4 spaces per level
3. Python checks conditions **top to bottom** — order matters!
4. Only **one block** runs in an `if-elif-else` chain — the first True one
5. `else` has **no condition** — it is always the last resort
6. Keep nesting **shallow** (max 2-3 levels) for readable code
