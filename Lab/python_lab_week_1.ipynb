## 1.1 📆 Display Current Date and Time

### 📖 Explanation
Python's built-in **`datetime`** module provides classes for working with dates and times.
The `datetime.now()` method returns the **current local date and time** as a `datetime` object.
We can then format it using **`strftime()`** (string format time) with format codes.

**Key `strftime()` format codes:**
```
%Y  -> 4-digit year          e.g. 2024
%m  -> 2-digit month         e.g. 07
%d  -> 2-digit day           e.g. 15
%H  -> Hour (24-hr)          e.g. 14
%I  -> Hour (12-hr)          e.g. 02
%M  -> Minutes               e.g. 35
%S  -> Seconds               e.g. 08
%A  -> Full weekday name     e.g. Monday
%B  -> Full month name       e.g. July
%p  -> AM or PM
```

### 🧮 Algorithm
```
Step 1: Import the datetime class from the datetime module
Step 2: Call datetime.now() to get current date and time
Step 3: Store the result in a variable
Step 4: Print the datetime object directly (default format)
Step 5: Use strftime() to display in custom readable formats
Step 6: Extract and display individual components (day, month, year, etc.)
```


````python
# Program 1.1: Display Current Date and Time

from datetime import datetime

# Step 1: Get the current date and time
now = datetime.now()

# Step 2: Display default format
print('=== Current Date and Time ===')
print(f'Default format   : {now}')

# Step 3: Custom formatted output
print(f'Date             : {now.strftime("%d-%m-%Y")}')
print(f'Time (24hr)      : {now.strftime("%H:%M:%S")}')
print(f'Time (12hr)      : {now.strftime("%I:%M:%S %p")}')
print(f'Full DateTime    : {now.strftime("%A, %d %B %Y | %I:%M:%S %p")}')

# Step 4: Extract individual components
print()
print('=== Individual Components ===')
print(f'Year             : {now.year}')
print(f'Month            : {now.month}  ({now.strftime("%B")})')
print(f'Day              : {now.day}')
print(f'Weekday          : {now.strftime("%A")}')
print(f'Hour             : {now.hour}')
print(f'Minute           : {now.minute}')
print(f'Second           : {now.second}')
print(f'Microsecond      : {now.microsecond}')
````

## 1.2 🐍 Get the Python Version

### 📖 Explanation
The **`sys`** module provides access to system-specific parameters and functions.
`sys.version` returns a string containing the Python version along with build details.
`sys.version_info` returns a named tuple with individual components (major, minor, micro).
The **`platform`** module gives additional environment details like OS and architecture.

**Key attributes:**
```
sys.version          -> Full version string
sys.version_info     -> NamedTuple (major, minor, micro, releaselevel, serial)
sys.executable       -> Path to Python interpreter
sys.platform         -> Operating system platform name
platform.python_implementation() -> CPython, PyPy, etc.
```

### 🧮 Algorithm
```
Step 1: Import the sys module
Step 2: Import the platform module
Step 3: Use sys.version to get and print the full version string
Step 4: Use sys.version_info to get major, minor, micro version numbers
Step 5: Display additional environment info using platform module
Step 6: Show the path to the Python interpreter using sys.executable
```


````python
# Program 1.2: Get Python Version

import sys
import platform

# Step 1: Full version string
print('=== Python Version Information ===')
print(f'Full Version     : {sys.version}')

# Step 2: Structured version info
info = sys.version_info
print(f'Major Version    : {info.major}')
print(f'Minor Version    : {info.minor}')
print(f'Micro (patch)    : {info.micro}')
print(f'Release Level    : {info.releaselevel}')
print(f'Version String   : {info.major}.{info.minor}.{info.micro}')

# Step 3: Platform and environment details
print()
print('=== Environment Details ===')
print(f'Operating System : {platform.system()} {platform.release()}')
print(f'Architecture     : {platform.machine()}')
print(f'Implementation   : {platform.python_implementation()}')
print(f'Interpreter Path : {sys.executable}')
print(f'Platform ID      : {sys.platform}')

# Step 4: Check version compatibility
print()
if sys.version_info >= (3, 8):
    print('✅ Python 3.8+ detected — f-strings, walrus operator supported!')
else:
    print('⚠️  Older Python version — consider upgrading.')
````

## 1.3 🔢 Compute n + nn + nnn

### 📖 Explanation
Given a single digit integer **n**, we need to compute:
- `n` = the digit itself (e.g., 5)
- `nn` = the digit repeated twice (e.g., 55)
- `nnn` = the digit repeated three times (e.g., 555)
- **Result** = n + nn + nnn

**How to build nn and nnn from n:**
```
n   =  5
nn  =  n * 10 + n          = 5*10 + 5  = 55
nnn =  n * 100 + n * 10 + n = 5*100 + 5*10 + 5 = 555

OR using string repetition:
nn  = int(str(n) * 2)   -> '5' * 2 = '55' -> int('55') = 55
nnn = int(str(n) * 3)   -> '5' * 3 = '555'-> int('555') = 555
```

**Result = n + nn + nnn = 5 + 55 + 555 = 615**

### 🧮 Algorithm
```
Step 1: Accept integer input n from the user
Step 2: Build nn by repeating the string representation of n twice,
        then convert back to int
Step 3: Build nnn by repeating the string representation of n three times,
        then convert back to int
Step 4: Compute result = n + nn + nnn
Step 5: Display n, nn, nnn and the final result
```


````python
# Program 1.3: Compute n + nn + nnn

# Step 1: Accept input
n = int(input('Enter a single digit integer n: '))

# Step 2: Build nn and nnn using string repetition
nn  = int(str(n) * 2)    # e.g. '5' * 2 = '55' -> 55
nnn = int(str(n) * 3)    # e.g. '5' * 3 = '555' -> 555

# Step 3: Compute result
result = n + nn + nnn

# Step 4: Display
print()
print(f'n         = {n}')
print(f'nn        = {nn}')
print(f'nnn       = {nnn}')
print(f'n+nn+nnn  = {n} + {nn} + {nnn} = {result}')

print()
# Alternative method using arithmetic
print('--- Alternative: Using Arithmetic ---')
nn_arith  = n * 10 + n
nnn_arith = n * 100 + n * 10 + n
print(f'nn  = n*10 + n        = {n}*10 + {n} = {nn_arith}')
print(f'nnn = n*100 + n*10 + n = {nnn_arith}')
print(f'Result (arithmetic)   = {n + nn_arith + nnn_arith}')

print()
# Show results for all single digits
print('--- Results for all single digits ---')
print(f'{"n":<5} {"nn":<7} {"nnn":<8} {"n+nn+nnn"}')
print('-' * 30)
for d in range(1, 10):
    d_nn  = int(str(d) * 2)
    d_nnn = int(str(d) * 3)
    print(f'{d:<5} {d_nn:<7} {d_nnn:<8} {d + d_nn + d_nnn}')
````

## 1.4 📦 Read and Print Various Types of Variables

### 📖 Explanation
Python supports several built-in data types. A **variable** is a name that refers to a value stored in memory.
Python is **dynamically typed** — you don't need to declare a type; Python infers it automatically.

This program demonstrates reading values of different types from the user and printing them
along with their type using **`type()`** and **`isinstance()`**.

**Types demonstrated:**
```
int      -> Whole numbers          42, -7
float    -> Decimal numbers        3.14, -0.5
complex  -> Complex numbers        2+3j
bool     -> True or False
str      -> Text                   'Hello'
list     -> Ordered collection     [1, 2, 3]
tuple    -> Immutable sequence     (1, 2, 3)
set      -> Unique elements        {1, 2, 3}
dict     -> Key-value pairs        {'a': 1}
NoneType -> Absence of value       None
```

### 🧮 Algorithm
```
Step 1: Declare variables of different data types
Step 2: Read integer, float and string from user using input()
        (Note: input() always returns str — convert as needed)
Step 3: Print each variable with its value and data type
Step 4: Show memory size of each variable using sys.getsizeof()
Step 5: Demonstrate type checking with isinstance()
```


````python
# Program 1.4: Read and Print Various Types of Variables

import sys

# ── Step 1: Read values from user ──
print('=== Enter Values ===')
int_val     = int(input('Enter an integer     : '))
float_val   = float(input('Enter a float        : '))
str_val     = input('Enter a string       : ')

# ── Step 2: Define other types ──
complex_val = complex(int_val, 3)               # e.g., 5+3j
bool_val    = int_val > 0                        # True if positive
list_val    = [int_val, float_val, str_val]      # mixed list
tuple_val   = (int_val, float_val)               # fixed pair
set_val     = {int_val, int_val + 1, int_val + 2}
dict_val    = {'name': str_val, 'score': int_val, 'gpa': float_val}
none_val    = None

# ── Step 3: Display all variables ──
print()
print('=== Variable Summary ===')
print(f'{"Variable":<15} {"Value":<30} {"Type":<12} {"Size (bytes)"}')
print('-' * 75)

all_vars = [
    ('int_val',     int_val),
    ('float_val',   float_val),
    ('complex_val', complex_val),
    ('bool_val',    bool_val),
    ('str_val',     str_val),
    ('list_val',    list_val),
    ('tuple_val',   tuple_val),
    ('set_val',     set_val),
    ('dict_val',    dict_val),
    ('none_val',    none_val),
]

for var_name, val in all_vars:
    val_str  = str(val)
    type_str = type(val).__name__
    size     = sys.getsizeof(val)
    print(f'{var_name:<15} {val_str:<30} {type_str:<12} {size}')

# ── Step 4: isinstance() checks ──
print()
print('=== isinstance() Type Checks ===')
print(f'int_val  is int   : {isinstance(int_val, int)}')
print(f'float_val is float: {isinstance(float_val, float)}')
print(f'str_val  is str   : {isinstance(str_val, str)}')
print(f'bool_val is bool  : {isinstance(bool_val, bool)}')
print(f'bool_val is int   : {isinstance(bool_val, int)}  (bool is subclass of int!)')
print(f'none_val is None  : {none_val is None}')
````

## 1.5 📅 Print Calendar of a Given Month and Year

### 📖 Explanation
Python's built-in **`calendar`** module provides functions for working with calendars.
`calendar.month(year, month)` returns a formatted multi-line string representing the calendar
for the specified month and year.

**Key functions in the `calendar` module:**
```
calendar.month(year, month)       -> String calendar for one month
calendar.calendar(year)           -> String calendar for entire year
calendar.monthrange(year, month)  -> (weekday of 1st day, total days)
calendar.isleap(year)             -> True if leap year
calendar.weekday(year, month, day)-> Weekday of a specific date (0=Mon)
calendar.setfirstweekday(0)       -> Set Monday as first day (0=Mon, 6=Sun)
```

### 🧮 Algorithm
```
Step 1: Import the calendar module
Step 2: Accept year (int) and month (int) from the user
Step 3: Validate month is between 1 and 12
Step 4: Use calendar.month(year, month) to get the calendar string
Step 5: Print the formatted calendar
Step 6: Display additional info: first weekday, total days, leap year check
```


````python
# Program 1.5: Print Calendar of a Given Month and Year

import calendar

# Step 1: Accept input from user
year  = int(input('Enter year  (e.g. 2024): '))
month = int(input('Enter month (1-12)    : '))

# Step 2: Validate month
if month < 1 or month > 12:
    print('❌ Invalid month! Please enter a value between 1 and 12.')
else:
    # Step 3: Print the calendar
    print()
    print(calendar.month(year, month))

    # Step 4: Additional information
    month_names = calendar.month_name  # ['', 'January', 'February', ...]
    day_names   = list(calendar.day_name)  # ['Monday', ..., 'Sunday']

    first_weekday, total_days = calendar.monthrange(year, month)

    print(f'Month Name    : {month_names[month]}')
    print(f'Year          : {year}')
    print(f'Total Days    : {total_days}')
    print(f'1st day falls : {day_names[first_weekday]}')
    print(f'Is Leap Year  : {calendar.isleap(year)}')

    # Show weekday of today in that month
    print()
    print(f'--- Weekdays for each date in {month_names[month]} {year} ---')
    for day in range(1, total_days + 1):
        weekday = calendar.weekday(year, month, day)
        marker = ' <-- 15th' if day == 15 else ''
        print(f'  {day:2d} -> {day_names[weekday]}{marker}')
````
