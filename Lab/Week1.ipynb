### 🐍 Python Program to Display Current Date and Time

### 📌 Program

```python
# Python program to display the current date and time

from datetime import datetime

now = datetime.now()

print("Current date and time:", now.strftime("%Y-%m-%d %H:%M:%S"))
```

---

### 📖 Explanation of the Program

This program displays the **current system date and time** using Python's built-in `datetime` module.

---

### 🔹 Step-by-Step Explanation

### 1️⃣ Importing the `datetime` Class

```python
from datetime import datetime
```

### Explanation:
- Python provides a built-in module called `datetime`.
- This module helps us work with:
  - dates
  - time
  - year
  - month
  - seconds
  - current time

### Here:
- `from datetime import datetime`
- imports the `datetime` class from the `datetime` module.

---

## 2️⃣ Getting the Current Date and Time

```python
now = datetime.now()
```

### Explanation:
- `datetime.now()` returns the current date and time from the system clock.
- The returned value is stored in the variable `now`.

### Example Value:
```python
2026-05-22 22:15:30.456789
```

This contains:
- Year
- Month
- Day
- Hour
- Minute
- Second
- Microseconds

---

## 3️⃣ Formatting the Date and Time

```python
now.strftime("%Y-%m-%d %H:%M:%S")
```

### Explanation:
- `strftime()` means:
  
  **String Format Time**

- It converts the date and time into a readable string format.

---

# 📅 Format Specifiers Used

| Specifier | Meaning | Example |
|---|---|---|
| `%Y` | 4-digit year | 2026 |
| `%m` | Month | 05 |
| `%d` | Day | 22 |
| `%H` | Hour (24-hour format) | 22 |
| `%M` | Minutes | 15 |
| `%S` | Seconds | 30 |

---

## 4️⃣ Displaying the Output

```python
print("Current date and time:", now.strftime("%Y-%m-%d %H:%M:%S"))
```

### Explanation:
- `print()` displays the formatted date and time on the screen.

---

# ✅ Sample Output

```python
Current date and time: 2026-05-22 22:15:30
```

---

# 🎯 Summary

This program:
1. Imports the `datetime` class
2. Gets the current system date and time
3. Formats it into a readable form
4. Displays the result

---



---

### 🔥 Alternative Formats

### Display Only Date

```python
print(now.strftime("%Y-%m-%d"))
```

### Output
```python
2026-05-22
```

---

### Display Only Time

```python
print(now.strftime("%H:%M:%S"))
```

### Output
```python
22:15:30
```

---

### ⚠️ Important Note

- `%H` → 24-hour format
- `%I` → 12-hour format

Example:

```python
print(now.strftime("%I:%M:%S %p"))
```

### Output
```python
10:15:30 PM
```
